```python
recordings_path = '/home/niell_lab/data/freely_moving_ephys/ephys_recordings/'
dates_path = ['020821_select']
search_data = '020821'
```

```python
collected_ephys = ephys_to_dataset(recordings_path, dates_path)
```

```python
top_dataset = xr.open_dataset(find(str(search_data)+'*TOP1*.nc',recordings_path)[0])
x_trace = top_dataset.TOP1_pts.sel(point_loc='tailbase_x').values
y_trace = top_dataset.TOP1_pts.sel(point_loc='tailbase_y').values
plt.plot(x_trace, y_trace,'k.',alpha=0.2)
```

```python
ephys_data = pd.read_json(find(str(search_data)+'*fm1_ephys_merge.json',recordings_path)[0])
goodcells = ephys_data.loc[ephys_data['group']=='good']
```

```python
n_units = len(goodcells)
plt.subplots(n_units,3, figsize=(10,100))
units = goodcells.index.values
panel_count = 1
unit_count = 1
for i, ind in enumerate(goodcells.index):
    unit_spikes = np.array(goodcells.at[units[i],'spikeT']) # spike times in seconds
    spike_positions = []
    for s in unit_spikes:
        pos_at_spike = [x_trace[int(s*60)], y_trace[int(s*60)]]
        spike_positions.append(pos_at_spike)
    spike_positions = np.array(spike_positions)
    num_bins_x = int(np.ceil(np.nanmax(x_trace) / 50))
    num_bins_y = int(np.ceil(np.nanmax(y_trace) / 50))
    spike_heatmap, spike_xedges, spike_yedges = np.histogram2d(np.nan_to_num(spike_positions[:,0], nan=0.0, posinf=0.0, neginf=0.0), np.nan_to_num(spike_positions[:,1], nan=0.0, posinf=0.0, neginf=0.0), bins=(num_bins_x, num_bins_y))
    total_heatmap, total_xedges, total_yedges = np.histogram2d(np.nan_to_num(x_trace, nan=0.0, posinf=0.0, neginf=0.0), np.nan_to_num(y_trace, nan=0.0, posinf=0.0, neginf=0.0), bins=(num_bins_x, num_bins_y))
    norm_heatmap = spike_heatmap / total_heatmap
    extent = [spike_xedges[0], spike_xedges[-1], spike_yedges[0], spike_yedges[-1]]
    
    plt.subplot(n_units,3,panel_count)
    plt.title('unit'+str(ind))
    plt.plot(x_trace, y_trace, '.', c='gray',alpha=0.1)
    plt.plot(spike_positions[:,0], spike_positions[:,1], '.', alpha=-0.2)
    plt.axis('off')
    panel_count = panel_count + 1
    
    plt.subplot(n_units,3,panel_count)
    norm_heatmap = np.nan_to_num(norm_heatmap, nan=0.0, posinf=0.0, neginf=0.0)
    # hx, hy = norm_heatmap.T.sum(axis=0), norm_heatmap.T.sum(axis=1)
    plt.imshow(norm_heatmap.T, extent=extent, origin='lower', cmap='inferno')
    plt.colorbar()
    plt.axis('off')
    panel_count = panel_count + 1
    
    plt.subplot(n_units,3,panel_count)
    data = collected_ephys.filter_by_attrs(stim='hf1_wn',unit='unit'+str(unit_count))
    unit_count = unit_count + 1
    try:
        plt.imshow(data.to_array().sel(ephys_params='sta').values.ravel()[0],vmin=-0.3,vmax=0.3,cmap='jet')
    except ValueError:
        plt.imshow(np.zeros([60,80]),cmap='jet')
    plt.axis('off')
    panel_count = panel_count + 1

plt.tight_layout()
plt.show()
```
