```python
import matplotlib.pyplot as plt
```

```python
import xarray as xr
```

```python
ncpath = r'T:\freely_moving_ephys\ephys_recordings\040221\Phil\IMUtest\040221_Phil_control_Rig2_IMUtest_imu.nc'
imu_data = xr.open_dataset(ncpath)
```

```python
imu_data.sample.values
```

```python
plt.subplots(4,1, figsize=(5,10))
plt.subplot(411)
for i in ['acc_x', 'acc_y', 'acc_z']:
    plt.plot(imu_data.to_array().sel(sample=i).T, label=i); plt.legend()
plt.subplot(412)
for i in ['gyro_x', 'gyro_y', 'gyro_z']:
    plt.plot(imu_data.to_array().sel(sample=i).T, label=i); plt.legend()
plt.subplot(413)
plt.plot(imu_data.to_array().sel(sample='roll').T)
plt.ylabel('roll')
plt.subplot(414)
plt.plot(imu_data.to_array().sel(sample='pitch').T)
plt.ylabel('pitch'); plt.xlabel('samples')
```
